```python
contents = [line.split(',') for line in open("ex1/ex1data1.txt").read().split()]
X = [float(element[0]) for element in contents]
Y = [float(element[1]) for element in contents]
X[:5], Y[:5], len(X), len(Y)
```

```python
def hypothesis_fn(theta, x):
    return theta[0] + theta[1] * x
```

```python
def error_fn(theta, x, y):
        return hypothesis_fn(theta, x) - y
```

```python
def mse(theta):
    m = len(X)
    
    return (1 / (2 * m)) * sum([
        error_fn(theta, X[i], Y[i]) ** 2 for i in range(m)
    ])

mse([0, 0])
```

```python
def ddx_mse(theta, term = 0):
    m = len(X)
    return (1 / m) * sum([
        error_fn(theta, X[i], Y[i]) if term == 0 
        else error_fn(theta, X[i], Y[i]) * X[i] 
        for i in range(m)
    ])
```

Have to beat J =  0.24552

```python
def gradient_descent(iterations):
    theta = [0, 0]
    
    for iteration in range(iterations):
        J = mse(theta)
        print(f"Cost: {J} at Iteration: {iteration}")
        
        alpha = 0.01
        const_term_diff = alpha * ddx_mse(theta, 0)
        var_term_diff = alpha * ddx_mse(theta, 1)
        
        theta[0] = theta[0] - const_term_diff
        theta[1] = theta[1] - var_term_diff
        
    return theta

gradient_descent(1500)

        
```
