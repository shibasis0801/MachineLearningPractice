```python
# small x represents one data point from X
def hypothesis_fn(theta, x):
    return sum([theta[i]*x[i] for i in range(len(theta))])
```

```python
# small x, small y are respective single samples
def error_fn(theta, x, y): 
    return hypothesis_fn(theta, x) - y
```

```python
def mse(theta, X, Y):
    m = len(X)
    return (1 / (2 * m)) * sum([
        error_fn(theta, X[i], Y[i]) ** 2 
        for i in range(m)
    ])
```

```python
def derivative_mse_theta(theta, X, Y, idx):
    m = len(X)
    return (1 / m) * sum([
        error_fn(theta, X[i], Y[i]) * X[i][idx]
        for i in range(m)
    ])
```
